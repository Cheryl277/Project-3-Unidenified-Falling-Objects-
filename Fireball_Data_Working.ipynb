```python
cleaned_df = df.loc[~(df["Latitude (deg.)"].isna())]
cleaned_df
```

```python
cleaned_df["lat"] = cleaned_df["Latitude (deg.)"].map(lambda l: parse(l))
```

```python
cleaned_df
```

```python
cleaned_df["long"] = cleaned_df["Longitude (deg.)"].map(lambda l: parse(l))
```

```python
cleaned_df
```

```python
geo = {"type":"FeatureCollection",
    "features":[]}
for idx, row in cleaned_df.iterrows():
    feat = {"type":"Feature",
           "properties":{
               "Impact Energy": row["Calculated Total Impact Energy (kt)"]
           },
           "geometry": {
               "type":"Point",
               "coordinates":[row["long"], row["lat"]]
            }}
    geo["features"].append(feat)
geo
```

```python
import json
```

```python
with open("example.geojson","w") as output:
    json.dump(geo,output)
```
